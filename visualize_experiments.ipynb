# Experimenting with nice ways to visualize things

```python
with open("examples.txt", "r") as f:
    data = f.readlines()

for i, d in enumerate(data):
    data[i] = d.strip().split(";")

for i in range(0, len(data), 2):
    data[i] = [int(x) for x in data[i]]
    data[i+1] = [float(x) for x in data[i+1]]
```

```python
import pprint

pprint.pprint(data)
```

```python
import math
import numpy as np
from hex import HexState

def visualize_game_board(gameboard):
    board_size = math.isqrt(len(gameboard))
    arr = list(np.array(gameboard).reshape(board_size, board_size))
    for i, sub_arr in enumerate(arr):
        arr[i] = list(sub_arr)
    gs = HexState.from_list(arr)
    gs.plot()
```

```python
visualize_game_board(data[6])
```

```python
import seaborn as sns
def visualize_probs(probs):
    board_size = math.isqrt(len(probs))
    sns.heatmap(np.array(probs).reshape(board_size, board_size))
```

```python
i = 18
visualize_game_board(data[i*2])
```

```python
visualize_probs(data[i*2+1])
```
